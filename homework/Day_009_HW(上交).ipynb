# 檢視與處理 Outliers
### 為何會有 outliers, 常見的 outlier 原因
* 未知值，隨意填補 (約定俗成的代入)，如年齡常見 0,999
* 可能的錯誤紀錄/手誤/系統性錯誤，如某本書在某筆訂單的銷售量 = 1000 本

### Hints: Emprical Cumulative Density Plot, [ECDF](https://zh.wikipedia.org/wiki/%E7%BB%8F%E9%AA%8C%E5%88%86%E5%B8%83%E5%87%BD%E6%95%B0), [ECDF with Python](https://stackoverflow.com/questions/14006520/ecdf-in-python-without-step-function)

```python
# 最大值離平均與中位數很遠
print(df['AMT_INCOME_TOTAL'].describe())

# 繪製 Empirical Cumulative Density Plot (ECDF)
"""
YOUR CODE HERE
"""
# sort the data:
data_sorted = np.sort(df['AMT_INCOME_TOTAL'])
# 計算數據的edcf值→ y
y = np.arange(1,len(data_sorted)+1)/len(data_sorted)


plt.plot(data_sorted, y)
plt.xlabel('Value')
plt.ylabel('ECDF')
#plt.xlim([y.index.min(), y.index.max() * 1.05]) # 限制顯示圖片的範圍
#plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()

# 改變 y 軸的 Scale, 讓我們可以正常檢視 ECDF
plt.plot(np.log(list(data_sorted)), y)
plt.xlabel('Value (log-scale)')
plt.ylabel('ECDF')

plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()
```

```python
# 最大值落在分布之外
print(df['REGION_POPULATION_RELATIVE'].describe())

# 繪製 Empirical Cumulative Density Plot (ECDF)
"""
Your Code Here
"""
# sort the data:
data_sorted2 = np.sort(df['REGION_POPULATION_RELATIVE'])

# 計算數據的edcf值→ y

y = np.arange(1,len(data_sorted2)+1)/len(data_sorted2)


plt.plot(data_sorted2, y)
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍
plt.show()

# 畫出直方圖
df['REGION_POPULATION_RELATIVE'].hist()
plt.show()

#計算出欄位中每個元素的個數
df['REGION_POPULATION_RELATIVE'].value_counts()

# 就以這個欄位來說，雖然有資料掉在分布以外，也不算異常，僅代表這間公司在稍微熱鬧的地區有的據點較少，
# 導致 region population relative 在少的部分較為密集，但在大的部分較為疏漏
```
