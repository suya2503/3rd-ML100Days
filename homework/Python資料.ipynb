## 字串:
* 字串相加
* k倍的字串
* 三引號(可以讓字串換行)
* 轉義字元\
* r:顯現原始字串，包含轉義字元

        * strip():針對某個字串，將其中的元素刪除

    * template = '{0:.2f} {1:s}are worth US${2:d}'.format(4.5560,'Argentine Pesos',1)

## 資料型態轉換
* 運用 type() 顯示資料型態
* int(), float(), str()
* 四捨五入與絕對值運算
    * round()：四捨五入轉整數運算
    * abs()：取絕對值 (Absolute value)

```python
print(3.14, int(3.14))
print(3.9999, int(3.9999))
print(3.0, int(3.0))
print(-3.999, int(-3.999))
print('2345', int('2345'))
print(int('23 瓶飲料'))
```

```python
print('2345', int('2345'))
#print(int('23 瓶飲料'))
print(float('123.45'))
print(type(float('123.45')))
print(float('123.45 元'))
```

```python
print(str(17))
print(str(123.45))
print(type(str(123.45)))
```

```python
print(type(str(123.45)))
print(3.14, round(3.14))
print(3.72, round(3.72))
print(3.756, round(3.756, 2))    # 小數 2 位
```

```python
print(abs(10-2), abs(2-10))
```

## 變數
* 程式中的各種名稱統稱為識別字 (Identifier)，包含模組名稱 (例如 convert.py) 、函式名稱 (例如 print()) 、或變數名稱 (例如 address) 等
* 變數名稱 命名規則:
    * 只能用英文、底線、數字
    * 不能以數字作為開頭
    * 不能使用Keyword作為一般識別字
    
    * 亦不要用內建的函式名稱做為一般識別字的名稱
    
    
* 識別字的英文字母大小寫有別 (Case-sensitive) ，因此 address, Address, aDDress, 與 ADDRESS 均為不同的識別字
* 
* 特性: 
    * 不需要事先宣告，利用指派指令直接創造新變數
    * 動態資料型別，只要指派指令表示式的資料型態改變，變數的資料型態就改變，例如以下變數 a 可以是整數、浮點數、或是字串
* 變數與記憶體:


* 型態強迫轉換 (Type coercion)
    * Python 不會強迫轉換變數型態

```python
52+'2'
```

```python
a = 3.14
b = 2
print(type(b))
c = a+b
print(type(a))
print(type(b))
print(type(c))
```

## 表示式
* 表示式是由運算元與運算子組成
* 


```python
#  /：一般除法，相除的結果一定是浮點數
x = 5
y = 2
print(x/y)
x = 6
y = 2
print(x/y)
```

```python
#  //：整除法，相除結果僅留整數部份 (如果兩個運算元均為整數，則結果是整數， 若某一個運算元是浮點數，則是浮點數)
x = 5
y = 2
print(x//y)
x = 5.0
y = 2
print(x//y)
```
